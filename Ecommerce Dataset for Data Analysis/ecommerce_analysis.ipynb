```python
import numpy as np
import pandas as pd
import plotly
```

```python
df = pd.read_csv(
    filepath_or_buffer="project1_df.csv"
)
```

```python
columns = df.columns
mapper = dict()
for column in columns:
    new_column = column.lower().replace(" ", "_")
    mapper[column] = new_column

df = df.rename(mapper, axis="columns")
```

```python
df['purchase_date'] = pd.to_datetime(df['purchase_date'])
```

```python
df.info()
```

```python
df.head()
```

```python
df['net_amount'] = np.round(df['net_amount'], decimals=2)
```

```python
df.head()
```

```python
gender_net_amount = (
    df
    .groupby('gender')
    .agg(net_amount=("net_amount", "sum"))
    .sort_values(by="net_amount", ascending=False)
)
```

```python
gender_net_amount
```

```python
product_category_net_amount = (
    df
    .groupby('product_category')
    .agg(net_amount=("net_amount", "sum"))
    .sort_values(by="net_amount", ascending=False)
)
```

```python
product_category_net_amount
```

```python
import plotly.express as px

# ---- Aggregate data ----
product_category_net_amount = (
    df
    .groupby('product_category', as_index=False)
    .agg(net_amount=("net_amount", "sum"))
    .sort_values(by="net_amount", ascending=True)  # ascending for horizontal bars
)

# ---- Plot ----
fig = px.bar(
    product_category_net_amount,
    x="net_amount",
    y="product_category",
    orientation="h",
    color_discrete_sequence=["#7EC8E3"],  # light blue
    title="Net Amount by Product Category",
    template="plotly_dark",
    text="net_amount"  # data labels
)

# ---- Layout tweaks ----
fig.update_traces(
    texttemplate="%{text:,.0f}",  # format numbers
    textposition="outside"
)

fig.update_layout(
    xaxis_title="Net Amount",
    yaxis_title="Product Category",
    showlegend=False,
    height=600,
    margin=dict(l=120, r=80, t=60, b=40)
)

fig.show()
```
